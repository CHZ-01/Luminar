```python
q1 = df["selling_price"].quantile(0.25)
q2 = df["selling_price"].quantile(0.50)
q3 = df["selling_price"].quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
print(q1,q2,q3,iqr,upper_bound,lower_bound,sep="\n")
```

```python
df[(df["selling_price"] > lower_bound) & (df["selling_price"] < upper_bound)]
```

```python
df["mileage"] = df["mileage"].str.strip(" kmplg/")
df["mileage"] = df["mileage"].astype(float)
df["mileage"]
```

```python
df["engine"] = df["engine"].str.strip(" C")
df["engine"] = df["engine"].astype(int)
df["engine"]
```

```python
df["max_power"] = df["max_power"].str.strip(" bhp")
df["max_power"] = df["max_power"].astype(float)
df["max_power"]
```

```python
df["torque"] = df["torque"].apply(func = lambda x : x.upper().split()[0])
df["torque"] = df["torque"].apply(func = lambda x : x.upper().split("N")[0])
df["torque"] = df["torque"].apply(func = lambda x : x.upper().split("K")[0])
df["torque"] = df["torque"].apply(func = lambda x : x.upper().split("@")[0])
df["torque"] = df["torque"].apply(func = lambda x : x.upper().split("(")[0])
df["torque"] = df["torque"].astype(float)
df["torque"]
```

```python
def price_category(price):
  if price < 1000000:
    return "Budget"
  elif price >= 1000000 and price < 3000000:
    return "Midrange"
  else:
    return "Luxury"
```

```python
df["price_category"] = df["selling_price"].apply(price_category)
df
```

```python
def mileage_category(mile):
  if mile <= 25:
    return "High Mileage"
  elif mile <= 15:
    return "Mid Mileage"
  else:
    return "Low Mileage"
```

```python
df["mileage_category"] = df["mileage"].apply(mileage_category)
df
```

```python
def company_name(name):
  return name.split()[0]
```

```python
# df["Company"] = df["Name"].apply(func = lambda x : x.split()[0])
df["Company"] = df["Name"].apply(company_name)
df
```
